# Line Plot

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
```

```python
x
```

```python
y = 3*x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

# Adding Title and Labels  & Changing Line Aesthetics

```python
plt.plot(x,y,color = 'r', linestyle = ':', linewidth = 1)
plt.title('X vs Y')
plt.xlabel('This is X axis')
plt.ylabel('This is y axis')
```

# Adding two lines in the same plot

```python
x
```

```python
y1 = 2*x
```

```python
y2 = 3*x
```

```python
plt.plot(x,y1, color ='r', linestyle = ':')
plt.plot(x,y2,color = 'y',linewidth = 4)
plt.grid(True)
plt.title("X Vs Y graph")
plt.xlabel("Year")
plt.ylabel("Production")
plt.show()
```

# Adding Sub Plots

```python
plt.subplot(1,3,1)
plt.plot(x,y1,color='r',linestyle=':',linewidth='2')
plt.subplot(1,3,2)
plt.plot(x,y2,color='g',linestyle='-.',linewidth='2')
plt.subplot(1,3,3)
plt.plot(x,y2,color='y',linestyle='-.',linewidth='2')
```


# Bar Plot

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
student = {'Bob':87,'guddu':33,'Matt':89,'Pappu':100}
```

```python
names = list(student.keys())
```

```python
names
```

```python
values = list(student.values())
```

```python
values
```

```python
plt.bar(names,values)
plt.show()
```

# Adding Title and Labels

```python
plt.bar(names,values)
plt.title('Distribution of student Marks')
plt.xlabel('Names of Students')
plt.ylabel('Marks Got')
plt.show()
```

# Horizontal Bar Plot

```python
plt.barh(names,values,color='g')
plt.title("Students mark sheet")
plt.xlabel('Names of students')
plt.ylabel('Marks obtained by students')
plt.show()
```

#  scatter-plot

```python
x = [4,2,5,8,9,2,7,9,3,6,5,3]
y = [2,9,3,4,43,3,2,4,5,9,4,8]
```

```python
plt.scatter(x,y)
plt.show()
```

# Changing Mark aesthetics

```python
plt.scatter(x,y,marker='+',c='r',s=100)
plt.show()
```

# Adding two marker in the same plot

```python
z = [1,2,3,4,5,4,4,4,4,4,9,8]
```

```python
plt.scatter(x,y,marker = '*', c='r', s = 100)
plt.scatter(x,z,c='g',s=200)
plt.show()
```

# Adding Sub - Plots

```python
plt.subplot(1,2,1)
plt.scatter(x,y, marker='+', c='r',s = 100)
plt.subplot(1,2,2)
plt.scatter(x,z,marker = '*', c='g',s=100)
plt.show()
```

# Pie-Chart

```python
fruit = ['Apple','Orange','mango','Guava']
quantity = [67,34,100, 29]
```

```python
plt.pie(quantity,labels = fruit)
plt.show()
```

# Changing Aesthetics

```python
plt.pie(quantity,labels = fruit, autopct = '%0.1f%%', colors = ['yellow','green','blue','red','pink'])
plt.show()
```

# DoughNut-Chart

```python
plt.pie(quantity,labels=fruit, radius =2)
plt.pie([1],colors = ['w'], radius=1)
plt.show()
```
